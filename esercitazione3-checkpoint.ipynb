#  QUAL E' IL MIGLIOR TEAM POKEMON?

Obiettivo: creare il miglior Team pokemon, 

```python
#### We import the datasets 
pokedex = pd.read_csv("pokemon_database.csv", encoding="latin-1")
region = pd.read_csv('pokemon_region.csv')
stats = pd.read_csv('pokemon_stats.csv')
```

```python
pokedex_new = pokedex[['Pokemon Id', 'Pokedex Number', 'Pokemon Name','Primary Type','Game(s) of Origin']].copy()
pokedex_new = pokedex_new.rename(columns={'Game(s) of Origin': "games"})      #.rename(columns={"nome_vecchio": "nome_nuovo"}))
stats_new = stats[['Pokedex Number','Health Stat', 'Attack Stat', 'Defense Stat','Speed Stat','Experience Growth Total']].copy()
df_new = pd.merge(pokedex_new,stats_new, on = 'Pokedex Number', how = 'outer')
df_new = pd.merge(df_new,region, on = 'games', how = 'outer')

```
