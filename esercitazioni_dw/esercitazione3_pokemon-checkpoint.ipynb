**COMBATTI CONTRO UN POKEMON:**

```python
# Our aim is to fight against a team. For each pokemon of their team, identified by their id  (pokemon_nemico_id),
# we choose the best one from the pokemon at our disposal (our_pokemons) according to a criterion (criterion).
## To this end we need to associate to each type its antagonist

#print(pokedex['Primary Type'].value_counts())
primary_antagonist = {
    "Water": "Electric",
    "Normal": "Fighting",
    "Grass": "Fire",
    "Bug": "Fire",
    "Psychic": "Dark",
    "Fire": "Water",
    "Electric": "Ground",
    "Rock": "Fighting",
    "Dark": "Fighting",
    "Fighting": "Psychic",
    "Ghost": "Dark",
    "Poison": "Ground",
    "Ground": "Water",
    "Dragon": "Ice",
    "Steel": "Ground",
    "Ice": "Fire",
    "Fairy": "Steel",
    "Flying": "Electric"
}
# This fuction returns a filtered dataframe with the pokemons that are strong against the pokemon with id pokemon_nemico_id, according to the primary type of the latter.
def scegli_pokemon(pokemon_nemico, our_pokemons = df_new, criterion = 'Attack Stat'): 
    # We get the primary type of the pokemon nemico
    primary_type_nemico = pokedex[pokedex['Pokemon Name'] == pokemon_nemico]['Primary Type'].values[0]
    #debug
    #print(primary_type_nemico)
    # We get the antagonist type
    antagonist_type = primary_antagonist[primary_type_nemico]
    #debug
    #print(antagonist_type)
    # We filter our pokemons according to the antagonist type
    filtered_pokemons = our_pokemons[our_pokemons['Primary Type'] == antagonist_type]
    if filtered_pokemons.empty:
        filtered_pokemons = our_pokemons
    #debug
    #print(filtered_pokemons)
    # We sort the filtered pokemons according to the criterion and we return the best one
    best_pokemon = filtered_pokemons.sort_values(by=criterion, ascending=False).iloc[0]
    return best_pokemon


```

**LOTTIAMO CONTRO I CAPO LEGA DI OGNI REGIONE**  
Costruiamo il nostro team perfetto per combattere i Capi lega 

```python
### Let's write a fuction that, given a team of pokemons, returns the optimal team to fight against it, according to our criterion.
def squadra_ottimale_contro_team(squadra_nemica, our_pokemons = df_new, criterion = 'Attack Stat'):
    squadra_ottimale = []
    for pokemon_nemico in squadra_nemica:
        best_pokemon = scegli_pokemon(pokemon_nemico, our_pokemons, criterion)
        squadra_ottimale.append(best_pokemon)
    return squadra_ottimale

```

```python
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO


def show_team_images(team_ids, pokedex_df):
    """
    Display images for a list of Pokmon IDs in one row (Series-free).
    
    Parameters:
        team_ids (list of int): Pokmon IDs
        pokedex_df (pd.DataFrame): Must have columns 'Pokemon Id' and 'Pokedex Number'
    """
    # Convert the pokedex to a dictionary for fast scalar lookup
    id_to_number = {int(row['Pokemon Id']): int(row['Pokedex Number']) 
                    for _, row in pokedex_df.iterrows()}
    
    num_pokemon = len(team_ids)
    fig, axes = plt.subplots(1, num_pokemon, figsize=(num_pokemon*3, 4))
    
    if num_pokemon == 1:
        axes = [axes]
    
    for ax, pokemon_id in zip(axes, team_ids):
        # Lookup Pokedex number as a plain scalar
        if pokemon_id not in id_to_number:
            raise ValueError(f"Pokmon ID {pokemon_id} not found in pokedex.")
        pokedex_number = id_to_number[pokemon_id]
        
        # Build URL
        url = f"https://www.serebii.net/xy/pokemon/{pokedex_number:03d}.png"
        
        # Download and display image
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        
        ax.imshow(img)
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

def names_to_ids(pokemon_names, pokedex_df):
    """
    Convert a list of Pokmon names to a list of IDs using the pokedex DataFrame.
    
    Parameters:
        pokemon_names (list of str): Pokmon names
        pokedex_df (pd.DataFrame): Must have columns 'Name' and 'ID'
        
    Returns:
        list of int: Corresponding Pokmon IDs
    """
    # Build a dictionary: name -> ID
    name_to_id = {row['Pokemon Name']: int(row['Pokemon Id']) for _, row in pokedex_df.iterrows()}
    
    # Convert names to IDs
    ids = []
    for name in pokemon_names:
        if name not in name_to_id:
            raise ValueError(f"Pokmon name '{name}' not found in pokedex.")
        ids.append(name_to_id[name])
    
    return ids
```
