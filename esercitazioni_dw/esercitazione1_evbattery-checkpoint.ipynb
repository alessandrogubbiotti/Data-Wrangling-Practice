### BATTERIE LFP vs NMC

1. Lithium Iron Phosphate -> batterie a litio 
2. Nickel Manganese Cobal -> batterie a nichel magnanese e cobalto

```python
df = pd.read_csv('ev_battery_degradation_v1.csv')
```

**PANORAMICA DELLE VARIABILI**
- Vehicle_ID: Identificatore unico per ogni veicolo.
- Car_Model: Modello specifico del veicolo elettrico (ad esempio, Tesla Model 3, BYD Atto 3).
- Battery_Type: Chimica elettrochimica della batteria (LFP o NMC).
- Battery_Capacity_kWh: Capacità totale di stoccaggio energetico della batteria.
- Vehicle_Age_Months: Tempo trascorso dalla produzione del veicolo (in mesi).
- Total_Charging_Cycles: Numero cumulativo di cicli completi di carica-scarica.
- Avg_Temperature_C: Temperatura media di funzionamento della batteria (cruciale per il degrado).
- Fast_Charge_Ratio: Proporzione di sessioni di ricarica effettuate tramite ricarica rapida DC (da 0,0 a 1,0).
- Avg_Discharge_Rate_C: Tasso medio di scarica dell’energia durante la guida
- Driving_Style: Etichetta categorica (Conservativo, Moderato, Aggressivo).
- Internal_Resistance_Ohm: Impedenza interna della batteria; aumenta quando lo stato di salute (SoH) diminuisce.
- SoH_Percent:  Stato di salute. Capacità attuale come percentuale della capacità originale.
- Battery_Status:  Etichetta di classificazione (Sana vs. Sostituire).

**OUTLIER**

```python
mean = df['Avg_Temperature_C'].mean()
std = df['Avg_Temperature_C'].std()

df['Z-score'] = (df['Avg_Temperature_C'] - mean) / std

outliers = df[abs(df['Z-score']) > 3]
print("Outlier identificati con Z-score > 3 o < -3:")
print(outliers[['Avg_Temperature_C', 'Z-score']])
```

```python
outlier_rows = df[abs(df['Z-score']) > 4]
outlier_rows
```

```python
mean = df['Battery_Capacity_kWh'].mean()
std = df['Battery_Capacity_kWh'].std()

df['Z-score'] = (df['Battery_Capacity_kWh'] - mean) / std

outliers = df[abs(df['Z-score']) > 3]
print("Outlier identificati con Z-score > 3 o < -3:")
print(outliers[['Battery_Capacity_kWh', 'Z-score']])
```

**Confronto statistico delle due batterie in base a Stato di Salute e di resistenza interna**  

a. State of Health (SoH): La batteria migliore risulta essere quella LFP, in quanto presenta una media più alta e una deviazione standard minore, indicando non solo uno stato di salute migliore ma anche maggiore uniformità tra le batterie.  
b. Internal Resistance (Ω): Anche per questa variabile la batteria migliore è quella LFP, poiché ha una media inferiore (ricordiamo che l’impedenza interna aumenta quando lo stato di salute diminuisce) e una deviazione standard più bassa, confermando stabilità e migliore condizione complessiva della batteria.

```python
tab_1 = (df.groupby("Battery_Type")["SoH_Percent"].agg(["mean", "median", "std"]))
tab_2 = (df.groupby("Battery_Type")["Internal_Resistance_Ohm"].agg(["mean", "median", "std"]))

print('--'*30)
print(f'Confronto tra le batterie in base allo stato di salute \n{tab_1}')
print('--'*30)
print(f'Confronto tra le batterie in base alla resistenza \n{tab_2}')
print('--'*30)
```

**Lo stile della guida influenza la stato di salute della batteria?**   
Uno stile di guida più aggressivo porterà a una degradazione maggiore della batteria indipendentemente dal tipo di questa. Tuttavia, si vede che le batterie LFP mantiene uno stato di salute e più alto rispetto alle NMC.

**Classificazione in base a Tempo e a Temperature**

```python
age_bins = [0, 12, 24, 36, 60, float("inf")]
age_labels = [
    "0–1 year",
    "1–2 years",
    "2–3 years",
    "3–5 years",
    "5+ years"
]

temp_bins = [-50, 0, 10, 25, 35, 60]
temp_labels = ['Cold', 'Cool', 'Moderate', 'Warm', 'Hot']

df["Age_Bin"] = pd.cut(
    df["Vehicle_Age_Months"],
    bins=age_bins,
    labels=age_labels,
    right=False
)


df['Temp_Bin'] = pd.cut(
    df['Avg_Temperature_C'],
    bins=temp_bins,
    labels=temp_labels,
    right=False
)
### Questo è per creare le categoriche
temp_counts = df.pivot_table(
    index="Battery_Type",       # rows
    columns="Temp_Bin",         # columns = the bin labels
    values="Avg_Temperature_C", # any numeric column works
    aggfunc="count",            # count of records
    fill_value=0                # fill empty bins with 0
)

# Optional: make column names exactly the labels
temp_counts.columns = temp_labels

temp_counts

age_counts = df.pivot_table(
    index="Battery_Type",       # rows
    columns="Age_Bin",         # columns = the bin labels
    values="Vehicle_Age_Months", # any numeric column works
    aggfunc="count",            # count of records
    fill_value=0                # fill empty bins with 0
)   
age_counts.columns = age_labels
age_counts

soh_temp = df.pivot_table(
    index="Battery_Type",       # rows = battery types
    columns="Temp_Bin",         # columns = the temperature bins
    values="SoH_Percent",       # aggregate SoH
    aggfunc="mean",             # average SoH in each bin
    fill_value=0                # fill missing bins with 0
)
soh_age = df.pivot_table(
    index="Battery_Type",
    columns="Age_Bin",
    values="SoH_Percent",
    aggfunc="mean",
    fill_value=0
)
soh_temp.columns = temp_labels
soh_age.columns = age_labels
soh_temp    
soh_age

import matplotlib.pyplot as plt

temp_counts_T = temp_counts.T
temp_counts_T.plot(kind='bar', stacked=True)
plt.title('Distribution of Temperature Bins by Battery Type')
plt.xlabel('Battery Type')
plt.ylabel('Count')
plt.legend(title='Temperature Bin')
plt.show()  

age_counts = age_counts.T
age_counts.plot(kind='bar', stacked=True)
plt.title('Distribution of Age Bins by Battery Type')
plt.xlabel('Battery Type')
plt.ylabel('Count')
plt.legend(title='Age Bin')
plt.show()  



soh_temp_T = soh_temp.T
soh_temp_T.plot(kind='bar')
plt.title('Average SoH by Temperature Bin and Battery Type')
plt.xlabel('Temperature Bin')
plt.ylabel('Average SoH (%)')
plt.legend(title='Battery Type')
plt.show()  

soh_age_T = soh_age.T
soh_age_T.plot(kind='bar')
plt.title('Average SoH by Age Bin and Battery Type')        
plt.xlabel('Age Bin')
plt.ylabel('Average SoH (%)')
plt.legend(title='Battery Type')
plt.show()
```
